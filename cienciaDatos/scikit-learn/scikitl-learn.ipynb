```python
import pandas as pd 
import numpy as np 
from sklearn.impute import SimpleImputer  #Esta clase nos sirve para trabajar con valores nulos
from sklearn.preprocessing import LabelEncoder #Esta clase solo funciona con la variable objetivo y solo es para problemas de clasificacion
from sklearn.preprocessing import OrdinalEncoder #Esta clase funciona con todas las demas columnas que requieran alguna conversion categorica inherente a un valor
from sklearn.preprocessing import StandardScaler #Esta clase nos permite escalar ciertas columnas o features o caracteristicas que tengan datos entre ellas muy elevados en comparacion de otras
from sklearn.preprocessing import OneHotEncoder #Esta clase tambien convierte columnas categoricas pero que no necesiteen un valor inherente 


```

```python
df = pd.read_csv('../../data/dogs.csv')
df 
```

```python
df.isnull().sum() #Nos muestra cuantos valores nulos hay en cada columna
```

```python
# si existe un valor nulo en la fila(axis = 0) puedo eliminarla si tengo al menos una(any) o puedo modificar
#con how = all si necesito que TODOS LOS REGISTRO deben tener N/A para ser eliminada.
#Con subset = ['A','B'] tiene que analizar especificamente estas columnas, al añadir este apartado se especifica cual columna se analiza, si no se especifica(NONE) es con todas
# con inplace = True devuelve el dataframe con las filas eliminadas.


df.dropna(axis=0, how='any', subset=None,inplace=True)
df
```

```python
df = pd.read_csv('../../data/dogs.csv')
df 
```

```python
imp = SimpleImputer(strategy='mean') #podemos tener varias estrategias pero la mas comun es un promedio o un valor mas frecuente

df[["Years", "Strength", "Height"]] = imp.fit_transform(df[["Years", "Strength", "Height"]])



```

```python
df
```

```python
df.isnull().sum()
```

```python
#Esta solo se aplica a mi columna de variables OBJETIVO y solamente para los problemas de Machine Learning de clasificación.
le = LabelEncoder()
df["Pedigree"] = le.fit_transform(df["Pedigree"])
```

```python
df
```

```python
imp = SimpleImputer(strategy='most_frequent')
df[["Breed"]] = imp.fit_transform(df[["Breed"]])
```

```python
df
```

```python
oe = OrdinalEncoder(dtype= int)
df[['Weight']] = oe.fit_transform(df[['Weight']])
```

```python
df
```

```python
ohe = OneHotEncoder(dtype=int, sparse_output=False , drop="first")
codificado = ohe.fit_transform(df[['Breed']])
```

```python
codificado
```

```python
df2 = pd.DataFrame(codificado,columns = ohe.get_feature_names_out(), index=df.index )
```

```python
df2
```

```python
df = pd.concat((df,df2),axis=1,)
```

```python
df
```

```python
df = df.drop("Breed", axis=1)
```

```python
df
```

```python
scaler = StandardScaler()
df[['Years', 'Strength', 'Height', 'Weight']] = scaler.fit_transform(df[['Years', 'Strength', 'Height', 'Weight']])
```

```python
df
```

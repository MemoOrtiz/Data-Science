```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('../../data/athlete_events.csv')
df
```

```python
#Usar la clausula de numpy para lo siguiente:
#Si la edad es > 26 aÃ±os, al peso restarle 2kg.

df['Weight'] = np.where(df['Age'] > 26, df['Weight'] - 2, df['Weight'])
df
```

```python
# Si la edad >  30 y NOC es USA, escribir "Distinguished" del nombre

df['Name'] = np.where((df['Age'] > 30) & (df['NOC'] == 'USA'), 'Distinguished ' +df['Name'],df['Name'])
df.sort_values(by='NOC', ascending=False)
df
```

```python
# Genere una funcion que remplace M -> MALE y F -> FEMALE
def remplazo(sex):
    if sex == 'M':
        return 'Male'
    elif sex == 'F':
        return 'Female'
    else:
        return sex  


df['Sex'] = df['Sex'].apply(remplazo)
df
```

```python
# Ahora con lambda haga quee el "Team" -> mayusculas
df['Team'] = df['Team'].apply(lambda x: x.upper())
df
```

```python
# Ordene el df por edad y nombre en orden descendiente (mayor a menor)
df = df.sort_values(by=['Age', 'Name'], ascending=[False, False])
df
```

```python
# Elimine columa ID y Event y genere un nuevo df
df_nuevo = df.drop(columns=['ID', 'Event'])
df_nuevo
```

```python
# Filtre columna por Name, Sex, Age, NOC, Year 
df_filtrado = df[['Name', 'Sex', 'Age', 'NOC', 'Year']]
df_filtrado
```

```python
# Filtre por renglones por edades de 21 a 23 y por aÃ±o de 2000 a 2002
df_8 = df[(df['Age'] >= 21) & (df['Age'] <= 23) & (df['Year'] >= 2000) & (df['Year'] <= 2002)]
df_8
```
